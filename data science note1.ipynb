## 1. data's attribute: 

### *Qualitative: They describe the characteristics of objects without giving the actual size or quantity.* 

nominal attribute: It is not intended to use these numbers quantitatively, which means mathematical operations are meaningless on nominal attributes. 

binary attribute: Symmetric and Asymmetric 

ordinal attribute: Possible values may have ranking, but the differences between are unknown.

### *Quantitative:* 

interval-scaled: Measured by equal unit scale 

ratio_scaled: Numerical attribute with inherent zero 


### *For interval scale properties (without zero), we cannot use one value to be a multiple of another value, but for ratio scales (with zero points) we can say that.*

## 2. Measurement of the central trend

(arithmetic) mean: The most common and effective numerical measure of the "center" of the data set. 
$$ \frac {\sum ^{N}_{i=1} {{x}_{i}}} {N} $$ 

* ***weights*** can be assigned before variables ：$$\frac {\sum ^{N}_{i=1} {{{w}_{i}x}_{i}}} {\sum ^{N}_{i=1} {{w}_{i}}}$$ 

* ***trimmed mean***: The trimmed mean is the mean after discarding the high and low extreme values, which eliminates the impact of a few extreme values. 

median: as for asymmetric data set, median is better to describe. 
* for odd numbers of data, median is the middle one, for the even, it is defined as  the average of the middle two values.
 
* Calculate the approximation of the median of the entire data set by ***interpolation***.
$$ {L}_{1}+(\frac {\frac {N} {2}+{(\sum  {freq})}_{l}} {{freq}_{median}})width $$ 
*where L1 denotes he lower bound of the median interval, N stands for the number of values in dataset, and there is also the frequency rate between all regions below the median interval, frequency(numbers of the value) of median interval, width(value of max() - min()) of the median interval.*
mode:
* A data set with two or more multiple numbers is ***multimodal***.
* For *asymmetric single-peak* numerical data, we have the following *empirical* relationship: $$ mean-mode\approx 3*(mean-median) $$ 

midrange: The mean of the largest and least values of the data set. 

* *left-skewed(Positive skew): mode > median > mean, while right skewed(Negative skew): mean < median < mode*.

## 3. Measurement data dispersion

range: `max() - min()`

quantile: a value that divides dataset into some *basically* equal-sized intervals.
* *basically* means there are maybe sometimes we can't divide them into perfect same-sized intervels.
* ***quartiles*** are the most common.
    * *Q1 Q2(median) Q3* are three quartiles respectively.
    * ***IQR*** is the  difference between Q3 and Q1, which means the scope covered by the middle half of the dataset: 
    $$ IQR={Q}_{3}-{Q}_{1} $$  

five-number summary: min(), Q1, Q2(median), Q3 and max()  


outliers: points of undue influence
* Usually find the value above the Q3 or below the Q1 at least 1.5 x IQR
* Removing this data will cause a change in the estimate of any parameter greater than a standard deviation.
* When cook's distance is bigger than 0.4 (*empirical value*).  
* boxplot: a plot shows quartiles by edge of the boxes, IQRs by the length of the boxes, median by the middle line in the boxes and the farthest point of two extended lines give out min() and max().

variance and standard deviation: The degree of dispersion of data distribution(defined only the "central" measurement is the mean)
* `std error = sqrt(var())`
* variance calculation formulas：
$$ Var(X)=\frac {1} {N}\sum ^{N}_{i=1} {{({x}_{i}-\mu{})}^{2}}=E[(X-\mu {)}^{2}] $$
* Standard deviation is a good indicator of the divergence of the data set. (*An observation generally does not stay away from the mean more than several times the standard deviation.*)
* Above dispersion is only used when central measurement is the *mean*.

## 4. Graphs of the basic statistical description of data

quantile plot: Horizontal axes denotes for data and Vertical axes denotes for f-value.
* f-value is a percentage means about f*100% of values are less then corresponding data.
    * about 0.25 for Q1, about 0.5 for Q2(median), about 0.75 for Q3(not exactly because maybe the dataset could not be devided perfectly)

quantile-quantile plot: We can observe whether there is a drift from one distribution to another.
* number of points decided by the little one
* the Q-Q plot can also check the ***normality*** in fitting-model feasibility check.
    * The data distribution can pass the check if it roughly conforms to the straight line.

histogram and frequency histogram:
* Buckets are equal-width
* Also can be used in checking the ***normality*** in fitting-model feasibility check.

scatter plot: Determining whether two variables exists relationships in the connection, modulus or trend, which is the most effective graphical method.
* Positive correlation, negative correlation and no correlation are reflected from the plot.
    * eg. the pattern of the point tilts from left to right, which means that the value of X increases with the decrease of Y, implying negative correlation.
* Covariance and Correlation Coefficient(normalized covariance) are statistics that measures the strength of a *linear relationship* between two variables.
    * can be simply used in measuring *nonlinear relationship* but the effect is very limited and it is better to use polynomial regression analysis and nonlinear regression analysis.
